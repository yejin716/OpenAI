```python
import torch

#1차원 텐서
torch.tensor([1,2,3,4])
```

```python
tensor = torch.tensor([1,2,3,4])

print('original tensor: ', tensor)
print('original tensor size', tensor.shape)
```

```python
#batch 차원 추가
tensor_unsqueeze = tensor.unsqueeze(dim=0)
print('batch_added tensor: ' ,tensor_unsqueeze)
print('batch_added tensor size: ', tensor_unsqueeze.shape)
#1차원 tensor >> 2차원 tensor
# 첫번째 차원 : 배치크기 (배치크기 : 1)
```

```python
from transformers import GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

input_text = '맛있네요'

#입력 텍스트 >> 인코딩(코드화) >> token_id list 변환
input_idx = tokenizer.encode(input_text)
print(input_idx)
```

```python
#token_id list >> tensor 변환

input_tensor = torch.LongTensor(input_idx)
print(input_tensor)
print(input_tensor.shape) # 1차원
```

```python
#batch 추가 >> 모델 입력 형태 변환

input_tensor_unsqueeze = input_tensor.unsqueeze(dim=0)
print(input_tensor_unsqueeze)
print(input_tensor_unsqueeze.shape) # 2차원

#입력텐서 [1,12]
#여기서 첫번째 차원 : 배치 크기(여기서는 1(dim=0), 두번째 차원 : sequence 길이 (여기서는 12))

# 특히 transformer 모델 계열에서는 positional encoding(포지셔널-위치 인코딩) 때문에 배치 차원이 필요함
```
