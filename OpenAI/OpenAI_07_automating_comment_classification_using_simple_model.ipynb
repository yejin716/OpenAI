```python
def simple_model(input_text):
    # ê¸ì •ì ì¸ í•µì‹¬ì–´ ë° ngram, ê°€ì¤‘ì¹˜
    positive_keywords = {
        "ìµœê³ ": 3,
        "ì¶”ì²œ": 3,
        "ì§„ì§œ ì¢‹ë‹¤": 2.5,
        "ëŒ€ë°•": 3,
        "ê·¹ì°¬": 3,
        "ì•„ëŠ‘": 2,
        "ì¢‹ì•˜ìŠµë‹ˆë‹¤": 2,
        "ë§›ì§‘": 3,
        "ì§„ë¦¬": 3,
        "í•«í”Œ": 2.5,
        "ë§›ìˆ": 2.5,
        "ì˜¤ê³  ì‹¶ë‹¤": 2.5,
        "ë„ˆë¬´ ì¹œì ˆí•´ì„œ": 2,
        "ê¸°ë¶„ ì¢‹ê²Œ": 1.5,
        "ë¶„ìœ„ê¸°ë„ ì¢‹ê³ ": 1.5,
        "ì§„ì§œ ã„¹ã…‡": 2,
        "ë§›ìˆì–´ìš”": 2,
        "ì¢‹ì€": 1,
        "ì™”ëŠ”ë°": 1,
        "ì •ë§": 1.5,
        "ì¢‹ì•„ìš”": 1
    }
    # ë¶€ì •ì ì¸ í•µì‹¬ì–´ ë° n-gramsì™€ ê°€ì¤‘ì¹˜
    negative_keywords = {
        "ì‹¤ë§": 3,
        "ì•„ì‰½ë‹¤": 2.5,
        "ê·¸ëƒ¥ ê·¸ë˜": 2.5,
        "ì‹¤ë§ìŠ¤ëŸ¬ì›€": 3,
        "ìŒì‹ì´ ë„ˆë¬´": 2,
        "ë„ˆë¬´ ì§œì„œ": 1.5,
        "ê·¸ëƒ¥ ê·¸ë¬ì–´ìš”": 2,
        "ê¸°ëŒ€ ì´í•˜ì˜€ìŠµë‹ˆë‹¤": 2,
        "ëŠ¦ê²Œ ë‚˜ì˜¤ê³ ": 1.5,
        "ë³„ë¡œì˜€ì–´ìš”": 2,
        "ì´ë ‡ê²Œ": 1,
        "ë„ˆë¬´": 1
    }

    #í‚¤ì›Œë“œ ê¸¸ì´ì— ë”°ë¼ ì •ë ¬(ë‚´ë¦¼ì°¨ìˆœ)
    sorted_pos_keywords = sorted(positive_keywords.keys(), key = len, reverse = True)
    sorted_neg_keywords = sorted(negative_keywords.keys(), key = len, reverse = True)

    positive_score = 0
    negative_score = 0

    # ê¸ì • í‚¤ì›Œë“œ ì ìˆ˜ ê³„ì‚°
    for key in sorted_pos_keywords:
        if key in input_text:
            positive_score = positive_keywords[key] #ì ìˆ˜ë§Œ ë½‘ê¸°
            input_text = input_text.replace(key, "")
            #ì¤‘ë³µ ì ìˆ˜ ë°©ì§€ ìœ„í•´ keywordë¥¼ textì—ì„œ ì œê±°
    # ë¶€ì • í‚¤ì›Œë“œ ì ìˆ˜ ê³„ì‚°
    for key in sorted_neg_keywords:
        if key in input_text:
            negative_score = negative_keywords[key] #ì ìˆ˜ë§Œ ë½‘ê¸°
            input_text = input_text.replace(key, "")
            #ì¤‘ë³µ ì ìˆ˜ ë°©ì§€ ìœ„í•´ keywordë¥¼ textì—ì„œ ì œê±°
    # ê²°ê³¼ ë°˜í™˜
    if positive_score > negative_score:
        return 1
    else:
        return 0
```

```python
import urllib.request
urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/tykimos/tykimos.github.io/master/warehouse/dataset/tarr_train.txt",
    filename="tarr_train.txt",
)

import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
def classify_text(input_text):
    output =  simple_model(input_text)
    return output
```

```python
path = '/content/tarr_train.txt'

df = pd.read_csv(path, delimiter='\t')
df.head()
```

```python
actual_labels = []
predicted_labels = []

for idx, row in df.iterrows():
    comment = row['comment']
    actual_label = row['label']
    predicted_label = classify_text(comment)

    actual_labels.append(actual_label)
    predicted_labels.append(predicted_label)

    print('comment: ', comment)
    print('actual label: ', actual_label)
    print('predicted label: ', predicted_label)
    print("--" * 50)
```

```python
#ì •í™•ë„ ê³„ì‚°

accuracy = accuracy_score(actual_labels, predicted_labels)

#y_real, y_pred ë¹„êµ (cost ì¤„ì´ëŠ”ê²Œ ìš°ë¦¬ì˜ ëª©í‘œ)

print(f'acc: {accuracy * 100: .2f}%')

print(set(actual_labels))
print(set(predicted_labels))
```

```python
#confusion_matrix ê³„ì‚°
cm = confusion_matrix(actual_labels, predicted_labels, labels=[1,0])

print("\nConfusion Matrix:")
print("         Predicted:")
print("         ê¸ì •    ë¶€ì •")
print("Actual")
print("ê¸ì •      {:<5}  {:<5}".format(cm[0][0], cm[0][1]))
print("ë¶€ì •      {:<5}  {:<5}".format(cm[1][0], cm[1][1]))
```
