```python
def simple_model(input_text):
    # 긍정적인 핵심어 및 ngram, 가중치
    positive_keywords = {
        "최고": 3,
        "추천": 3,
        "진짜 좋다": 2.5,
        "대박": 3,
        "극찬": 3,
        "아늑": 2,
        "좋았습니다": 2,
        "맛집": 3,
        "진리": 3,
        "핫플": 2.5,
        "맛있": 2.5,
        "오고 싶다": 2.5,
        "너무 친절해서": 2,
        "기분 좋게": 1.5,
        "분위기도 좋고": 1.5,
        "진짜 ㄹㅇ": 2,
        "맛있어요": 2,
        "좋은": 1,
        "왔는데": 1,
        "정말": 1.5,
        "좋아요": 1
    }
    # 부정적인 핵심어 및 n-grams와 가중치
    negative_keywords = {
        "실망": 3,
        "아쉽다": 2.5,
        "그냥 그래": 2.5,
        "실망스러움": 3,
        "음식이 너무": 2,
        "너무 짜서": 1.5,
        "그냥 그랬어요": 2,
        "기대 이하였습니다": 2,
        "늦게 나오고": 1.5,
        "별로였어요": 2,
        "이렇게": 1,
        "너무": 1
    }

    #키워드 길이에 따라 정렬(내림차순)
    sorted_pos_keywords = sorted(positive_keywords.keys(), key = len, reverse = True)
    sorted_neg_keywords = sorted(negative_keywords.keys(), key = len, reverse = True)

    positive_score = 0
    negative_score = 0

    # 긍정 키워드 점수 계산
    for key in sorted_pos_keywords:
        if key in input_text:
            positive_score = positive_keywords[key] #점수만 뽑기
            input_text = input_text.replace(key, "")
            #중복 점수 방지 위해 keyword를 text에서 제거
    # 부정 키워드 점수 계산
    for key in sorted_neg_keywords:
        if key in input_text:
            negative_score = negative_keywords[key] #점수만 뽑기
            input_text = input_text.replace(key, "")
            #중복 점수 방지 위해 keyword를 text에서 제거
    # 결과 반환
    if positive_score > negative_score:
        return 1
    else:
        return 0
```

```python
import urllib.request
urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/tykimos/tykimos.github.io/master/warehouse/dataset/tarr_train.txt",
    filename="tarr_train.txt",
)

import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
def classify_text(input_text):
    output =  simple_model(input_text)
    return output
```

```python
path = '/content/tarr_train.txt'

df = pd.read_csv(path, delimiter='\t')
df.head()
```

```python
actual_labels = []
predicted_labels = []

for idx, row in df.iterrows():
    comment = row['comment']
    actual_label = row['label']
    predicted_label = classify_text(comment)

    actual_labels.append(actual_label)
    predicted_labels.append(predicted_label)

    print('comment: ', comment)
    print('actual label: ', actual_label)
    print('predicted label: ', predicted_label)
    print("--" * 50)
```

```python
#정확도 계산

accuracy = accuracy_score(actual_labels, predicted_labels)

#y_real, y_pred 비교 (cost 줄이는게 우리의 목표)

print(f'acc: {accuracy * 100: .2f}%')

print(set(actual_labels))
print(set(predicted_labels))
```

```python
#confusion_matrix 계산
cm = confusion_matrix(actual_labels, predicted_labels, labels=[1,0])

print("\nConfusion Matrix:")
print("         Predicted:")
print("         긍정    부정")
print("Actual")
print("긍정      {:<5}  {:<5}".format(cm[0][0], cm[0][1]))
print("부정      {:<5}  {:<5}".format(cm[1][0], cm[1][1]))
```
