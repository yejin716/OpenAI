```python
from transformers import pipeline

pipe = pipeline('sentiment-analysis', model="sangrimlee/bert-base-multilingual-cased-nsmc")

def dl_model(input_text):
    output = pipe(input_text)

    #예측 결과 첫번째 레이블 감성분석
    if output[0]['label'] == 'positive':
        return 1
    else:
        return 0



```

```python
import urllib.request

urllib.request.urlretrieve("https://raw.githubusercontent.com/tykimos/tykimos.github.io/master/warehouse/dataset/tarr_train.txt",
    filename="tarr_train.txt")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

def classify_text(text):
    output = dl_model(text)
    return output


#파일을 데이터프레임 로드
path = '/content/tarr_train.txt'
raw = pd.read_csv(path, delimiter='\t') # delimiter 구분자
df = raw.copy()
```

```python
df.head()
```

```python
actual_labels = []
predicted_labels = []

for index, row in df.iterrows():
    comment = row['comment']
    actual_label = row['label']
    predicted_label = classify_text(comment)

    actual_labels.append(actual_label)
    predicted_labels.append(predicted_label)

    if actual_label != predicted_label:
        print(f'Comment: {comment} Actual Label: {actual_label} Predicted Label: {predicted_label}\n')
        print('--'*100)

```

```python
#정확도 계산

accuracy = accuracy_score(actual_labels, predicted_labels)
print(f'acc: {accuracy * 100: .2f}%')

print(set(actual_labels))
print(set(predicted_labels))
```

```python
# confusion metrix(오분류)

cm = confusion_matrix(actual_labels, predicted_labels, labels=[1,0])

print("\nConfusion Matrix:")
print("         Predicted:")
print("         긍정    부정")
print("Actual")
print("긍정      {:<5}  {:<5}".format(cm[0][0], cm[0][1]))
print("부정      {:<5}  {:<5}".format(cm[1][0], cm[1][1]))
```
