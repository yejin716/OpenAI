Zero-shot Classification

```python
from transformers import pipeline

classifier = pipeline('zero-shot-classification')

classifier(
    "This is a course about tje generative AI library",
    candidate_labels= ['education', 'politics', 'business']


)
```

```python
from transformers import pipeline
import gradio as gr

classifier = pipeline("zero-shot-classification")

def classify_text(text, labels):
    labels = labels.split(",")
    result = classifier(text, candidate_labels=labels)

    # 결과 딕셔너리 생성
    result_dict = {}
    for i in range(len(result['labels'])):
        result_dict[result['labels'][i]] = result['scores'][i]

    return result_dict


iface = gr.Interface(
    fn=classify_text,
    inputs = [
        gr.Textbox(label='Text to classify'),
        gr.Textbox(label='Candidate Labels')
    ],
    outputs=gr.Label(label='classification Results', num_top_classes=3)
)

iface.launch(share=True)
```

개체명 (NER : Named Entity Recognition)

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Yellee and I work at MultiCampus in Busan")
```

```python
from transformers import pipeline
import gradio as gr

# NER pipeline 초기화
ner = pipeline("ner", grouped_entities=True)

def extract_entities(text):
    result = ner(text)
    #코드 수정
    entities = {}
    for entity in result:
        entities[entity['word']] = [entity['entity_group'], entity['score']]

    return entities

text = "My name is Yellee and I live at Busan"
print(extract_entities(text))

iface = gr.Interface(
    fn = extract_entities,
    inputs = gr.Textbox(label = 'Text for NER'),
    outputs = gr.Textbox(label = 'result')
)

iface.launch(share=True)
```

질의응답 (QnA)

```python
from transformers import pipeline
question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do you work?",
    context="My name is emma and I work at Mulcam in Seoul.",
)
```

```python
from transformers import pipeline
import gradio as gr

question_answerer = pipeline("question-answering")

def answer_qa(question, context):
    result = question_answerer(question=question, context=context)
    return (result['answer'], result['score'])


iface = gr.Interface(
    fn = answer_qa,
    inputs = [
        gr.Textbox(label = 'Question'),
        gr.Textbox(label = 'Context')
    ],
    outputs = [
        gr.Textbox(label = 'Answer'),
        gr.Textbox(label = 'Confidence')]

)
iface.launch(share=True)
```
