```python
#펠린드롬 ex) 스위스, 기러기, 토마토
# 뒤집어도 같은 문자

import gradio as gr

def flip_text(x):
    return x[::-1]

with gr.Blocks()as blocks_test:
    gr.Markdown("뒤집고 싶은 문자열을 입력하고, **실행**을 눌러 결과 확인하세요")
    with gr.Row():
        inp = gr.Textbox(placeholder='뒤집고 싶은 문자열')
        out = gr.Textbox()
    btn = gr.Button("실행")
    btn.click(fn=flip_text, inputs=inp, outputs=out)
blocks_test.launch()
```

펠린드롬 검사기 만들기

```python
import gradio as gr

def gr_is_palindrome(text):
    reverse_text = text[::-1]
    if text == reverse_text:
        return (reverse_text, '펠린드롬입니다.')
    else:
        return (reverse_text, '펠린드롬이 아닙니다.')

interface = gr.Interface(
    fn=gr_is_palindrome,
    inputs='text',
    outputs=[gr.Textbox(label="결과 단어"), gr.Textbox(label="판정 결과")]
)

interface.launch()
```
