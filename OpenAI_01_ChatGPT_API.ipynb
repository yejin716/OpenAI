패키지 불러오기

```python
import openai
```

Client 생성 및 API Key 지정하기

```python
#client= openai.OpenAI(api_key = 'api key 입력해주세요')
```

```python
client = openai.OpenAI(api_key = '')
```

기본 질문하기

```python
response = \
client.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [{ 'role':'user', 'content' : 'Tell me how to make a pizza'}]

)
```

```python
print(response)
```

응답에서 답변만 추출하기

```python
response.choices[0]
```

```python
response.choices[0].message
```

```python
response.choices[0].message.content
```

```python
print(response.choices[0].message.content)
```

소모 토큰, 비용확인

```python
print(response.usage)
```

```python
total_bill = response.usage.prompt_tokens * 0.0015/1000 + response.usage.completion_tokens * 0.002/1000
total_bill
```

```python
print('총 발생 비용{} 원'.format(total_bill * 1,323.43))
```

역할 부여하기

```python
response = \
client.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [{ 'role' : 'system', 'content' : '너는 친절하게 답변해 주는 비서야'}, #역할 부여
                { 'role' : 'user', 'content' : '2020년 월드시리즈에서 누가 우승했어?'} #사용자 질문

]

)

print(response.choices[0].message.content)
```

```python
response = \
client.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [{"role": "system", "content": "You must only answer users' questions in English. This must be honored. You must only answer in English."},
                { 'role' : 'user', 'content' : '2020년 월드시리즈에서 누가 우승했어?'}

]

)

print(response.choices[0].message.content)
```

```python
response = \
client.chat.completions.create(
    model = 'gpt-3.5-turbo',
    messages = [{"role": "system", "content": "You're a chatbot that refuses to answer and says sorry when users ask questions."},
                { 'role' : 'user', 'content' : '2020년 월드시리즈에서 누가 우승했어?'}

]

)

print(response.choices[0].message.content)
```

```python
response = client.chat.completions.create(
 model="gpt-3.5-turbo",
 messages=[
  {"role": "system", "content": "You are a translator that translates users' inputs. If the input is in Korean, it must be translated into English. This must be strictly adhered to."}, #번역
  {"role": "user", "content": "2020년 월드시리즈에서는 누가 우승했어?"}
  ]
 )
print(response.choices[0].message.content)
```

이전 대화를 포함하여 답변하기
